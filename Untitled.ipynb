```python
def effect_space_distance(s, filt):
    diffs = []
    fstd = []
    pstd = []
    for row in s.index:
        if np.all([name in row['group'] for name in filt]):
            diffs.append(np.abs(row['reference'][:,None] - row['final'].T))
            fstd.append(np.std(row['final'].T, axis=1))
            pstd.append(row['std'])
    diffs = np.hstack(diffs)
    fstd = np.hstack(fstd)
    stds = np.hstack(stds)
    return (np.tensordot(diffs, 1/s.sigmata, (0,0)),
            np.dot(np.mean(fstd, axis=1), 1/s.sigmata), # mean_across_fits(std_in_fit(final))
```

```python
def find_name(names, string):
    for name in names:
        if name in string:
            return name

val = []
for s in S:
    for gname in s.gnames_raw:
        gen = find_name(['cluster','bubble'], gname)
        fit = find_name(['DE', 'xGA', 'mGA'], gname)
        mut = find_name(['unweighted', 'graded', 'target-only'], gname)
        noisy = find_name(['clean', 'noisy'], gname)
        rand = find_name(['base', 'rand'], gname)

        val.extend([{'gen':gen,
                     'fit':fit,
                     'mut':mut,
                     'noisy':noisy,
                     'rand':rand,
                     'model': s.modelname,
                     'norm_percent_std': row['norm_percent_std'][i]
                    }
                    for row in s.index_by_group(gname)
                    for i in range(row['n_subpops'])
                   ])
df = pd.DataFrame(val, columns = ['gen', 'fit', 'mut', 'noisy', 'rand', 'model'])
```

```python
def filtered_boxplot(measure, filts = []):
    data = np.array([[np.ravel([row[measure] for row in s.index_by_group(gname)])
                      for gname in s.gnames if np.all([f in gname for f in filts])] for s in S])
    labels = [gname for gname in S[0].gnames if np.all([f in gname for f in filts])]
    plt.figure(figsize=(6,8))
    fa.boxplot(np.moveaxis(data, 0, 2), labels, [s.modelname for s in S])
```
